```python
struc1 = np.array([[1,1], [1,1]])
struc2 = np.array([[True, True]])
struc3 = np.ones((5,5))

struc = [struc1,struc2,struc3]
```

```python
fig, ax = plt.subplots(3,2, figsize=(10,10))
ax= ax.flatten()
for j in range(len(struc)): 
    ax[2*j].imshow(pattern1,cmap='gray', vmin=0, vmax=1)
    if j == 0 or j==1:
        dilated = ndimage.binary_dilation(pattern1, struc[j], origin=(0,0))
    else:
        dilated = ndimage.binary_dilation(pattern1, struc[j])
    # print(dilated)
    ax[2*j+1].imshow(dilated,cmap='gray', vmin=0, vmax=1)

ax[0].set_title('Original')
ax[1].set_title('Dilated')
```

```python
plt.figure()
plt.imshow(malaria, cmap='gray')
plt.figure()
plt.plot(np.histogram(malaria, bins=256)[1][:-1], np.histogram(malaria, bins=256, density=False)[0]) #Displays grayscale histogram of image
```

```python
def gray_thresh(image, low, high):
    image[image<low] = 0 
    image[image>high] = 255 
    return image
thresholded= gray_thresh(malaria, 121, 164)
plt.figure()
plt.imshow(thresholded, cmap='gray')
val= filters.threshold_otsu(thresholded[np.isfinite(thresholded)])
plt.figure()
BW = thresholded > val
plt.imshow(BW, cmap='gray')
```

```python
strel1 = np.ones((3,3))
plt.figure()
plt.imshow(BW, cmap='gray')
BW2= ndimage.binary_opening(BW, strel1)
plt.figure()
plt.imshow(BW2, cmap='gray')
BW3 = ndimage.binary_closing(BW2, strel1)
plt.figure()
plt.imshow(BW3, cmap='gray')
```
