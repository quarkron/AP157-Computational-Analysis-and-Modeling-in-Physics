## Creating Synthetic Images

```python
N = 500 #the higher, the finer
x = np.linspace(-1,1,num = N)
y= np.linspace(-1,1,num=N)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2) #Setup a circle, centered at the origin.
A = np.zeros(np.shape(R)) 
A[np.where(R<0.5)] = 1.0 #Circle of radius 0.5

#display as a white circle in 2D coordinates
plt.figure(figsize=(10,10))
plt.imshow(A, cmap = "Reds")

#display as a 3D surface in Cartesian coordinates
fig = plt.figure(figsize=(10,10)) 
ax = Axes3D(fig)
ax.plot_surface(X,Y,A, cmap='Reds')
```

```python
N = 500 # The higher N, the finer the resolution
f = 4 # Frequency
x = np.linspace(-2,2,num = N)
y = x 
X,Y = np.meshgrid(x,y) 

omega = 2*np.pi*f #Angular frequency
vals = (np.sin(omega*X)+1)*255/2 #Equation for sinusoid, with range 0 to 255
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)
ax.plot_surface(X,Y,vals, cmap='viridis')
ax.view_init(-160, 80)
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('y', fontsize=16)
ax.set_zlabel('z', fontsize=16)
plt.savefig('sinusoida.png', dpi=300)
```

## Synthetic colored images

```python
def hexagon_plotter(c, col='#FFD700'):
    s = 1 
    s1 = s/2 
    s2 = s*(np.sqrt(3))/2
    v1, v2, v3, v4, v5, v6 = (c[0] + s1, c[1] + s2), (c[0]+s, c[1]), (c[0] + s1, c[1] - s2), (c[0] - s1, c[1] - s2), (c[0]-s, c[1]), (c[0] - s1, c[1] + s2)
    polygon1 = Polygon([v1,v2,v3,v4,v5,v6,])
    polygon1.set_facecolor(col)
    polygon1.set_edgecolor('k')
    ax.add_patch(polygon1)
    plt.ylim(0,12)
    plt.xlim(0,12)

def hex_centers(c): 
    s = 1 
    s1 = 1.5*s
    s2 = s*np.sqrt(3)/2
    v1 = (c[0] + s1, c[1] + s2) 
    v2 = (c[0] + s1, c[1] - s2) 
    v3 = (c[0], c[1] - s2*2) 
    v4 = (c[0] - s1, c[1] - s2) 
    v5 = (c[0] - s1, c[1] + s2) 
    v6 = (c[0], c[1] + s2*2) 
    return (v1,v2,v3,v4,v5,v6)

```

```python
c = (6,6)
fig, ax = plt.subplots(1,1, figsize=(12,12))
hexagon_plotter(c)
a = hex_centers(c)
for center in a:
    hexagon_plotter(center)
    tier2 = hex_centers(center)
    for center2 in tier2: 
        hexagon_plotter(center2)
hexagon_plotter(c, col='w')
plt.xticks([])
plt.yticks([])
plt.savefig('jwst.jpg', dpi=300)
plt.savefig('jwst.png', dpi=300) 
plt.savefig('jwst.tiff', dpi=300) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 500
x = np.linspace(-10,10,num = N)
y = x
X,Y = np.meshgrid(x,y)
Rd, Gn, Bl = np.zeros((N,N)), np.zeros((N,N)), np.zeros((N,N))

#draw colored circles
Rt, Rc, deg = 3, 4, 30
xt, yt = Rt*np.cos(deg*np.pi/180), Rt*np.sin(deg*np.pi/180)

R = np.sqrt((X)**2 + (Y+Rt)**2)
Rd[np.where(R<Rc)]=1.0

R = np.sqrt((X-xt)**2 + (Y-yt)**2)
Gn[np.where(R<Rc)]=1.0

R = np.sqrt((X+xt)**2 + (Y-yt)**2)
Bl[np.where(R<Rc)]=1.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure(figsize=(10,10))
plt.imshow(I)
```

```python
#Olympic rings, subtractive color mixing

N=1000 
x= np.linspace(0,10,num=N)
y=np.linspace(0,30,num=3*N)
X,Y = np.meshgrid(x,y) 
def circle_plotter(r_o, x,y, c, r_i, Rd, Gn, Bl): 
    R = np.sqrt((X-x)**2 + (Y-y)**2) 
    reg = (R<r_o)*(R>r_i)
    if c == 0: #blue circle
        Rd[np.where(reg)] = 0.0 
        Gn[np.where(reg)] = 0.0 
    if c == 1: #yellow circle
        Bl[np.where(reg)] = 0.0 
    if c == 2: #black circle 
        Rd[np.where(reg)] = 0.0 
        Gn[np.where(reg)] = 0.0  
        Bl[np.where(reg)] = 0.0 
    if c == 3: #green circle
        Rd[np.where(reg)] = 0.0 
        Bl[np.where(reg)] = 0.0 
    if c==4: #red circle
        Gn[np.where(reg)] = 0.0  
        Bl[np.where(reg)] = 0.0 
    return (Rd, Gn, Bl)

I = np.ones((N,N,3))
l, s = 1.5, 2
c = [2,5]
c1 = [c[0]+l, c[1] + np.sqrt(s**2 - l**2)] 
c2= [c[0]+2*l, c[1]]
c3= [c[0]+3*l, c[1]+np.sqrt(s**2 - l**2)] 
c4=[c[0]+4*l, c[1]]
C=[c,c1,c2,c3,c4]
R_o, R_i = 0.4*(s+l),  0.4*(s+l)-0.2
r,g,b = np.ones((N,N)), np.ones((N,N)), np.ones((N,N))
for col in range(5):
    a=circle_plotter(R_o, x=C[col][0], y = C[col][1], r_i = R_i, c=col, Rd=r, Gn=g, Bl=b)
    r,g,b = a[0], a[1], a[2] 

I = np.zeros((N,N,3))
I[...,0] = r
I[...,1] = g
I[...,2] = b


fig = plt.figure(figsize=(10,10))
plt.imshow(I)
# plt.savefig('olympic rings.png', dpi=300) 
```
